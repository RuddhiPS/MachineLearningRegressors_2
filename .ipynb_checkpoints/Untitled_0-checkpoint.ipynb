## Question 2:

```python
# Initialize gradients for stochastic calculation
grad_theta0_sgd = 0.0
grad_theta1_sgd = 0.0

theta0 = torch.autograd.Variable(torch.tensor([0.0]), requires_grad=True) # we can take any initial values of theta0 and theta1 as mentioned in the question
theta1 = torch.autograd.Variable(torch.tensor([0.0]), requires_grad=True)

for i in range(num_samples):
    # For each data point
    xi = x[i]
    yi = y[i]
    
    # Forward pass
    y_pred = model(xi)
    
    # Loss function
    loss_i = (y_pred - yi) ** 2
    
    # Backward pass
    loss_i.backward()
    
    print("Data Point", i," : ", theta0.grad.item(), theta1.grad.item(), loss_i)
    
    # Accumulate gradients
    grad_theta0_sgd += theta0.grad.item()
    grad_theta1_sgd += theta1.grad.item()
    
    # Zero out gradients after each step
    theta0.grad.zero_()
    theta1.grad.zero_()

# Calculate the average gradients
grad_theta0_sgd /= num_samples
grad_theta1_sgd /= num_samples

print(f"Stochastic Gradient Average: theta0 = {grad_theta0_sgd}, theta1 = {grad_theta1_sgd}")
```

```python
print(f"True Gradient: theta0 = {true_grad_theta0}, theta1 = {true_grad_theta1}")
print(f"Stochastic Gradient Average: theta0 = {grad_theta0_sgd}, theta1 = {grad_theta1_sgd}")
```

Conclusion:

The true gradient is computed over the entire dataset in batch gradient descent. It gives the exact direction to move in parameter space (i.e., how to update the parameters) to minimize the loss function.

In stochastic gradient descent (SGD), instead of computing the gradient based on the whole dataset, you compute the gradient based on a single data point or a small batch of data.

We observe:
In expectation, the stochastic gradient is an unbiased estimate of the true gradient, i.e. on average, the stochastic gradient (calculated using individual data points) will be equal to the true gradient (calculated using the entire dataset)

Hence, though the individual stochastic gradients are noisy and seem inaccurate, on average, they point in the same direction as the true gradient.

Thus, we can say that SGD is noisy, but efficient, in terms of computation as well as space.

```python
# Helper function to initialize parameters
def initialize_params():
    theta0 = torch.tensor([0.0], requires_grad=True)  # Intercept
    theta1 = torch.tensor([0.0], requires_grad=True)  # Slope
    return theta0, theta1
```

```python
# Linear regression model
def model(x, theta0, theta1):
    return theta1 * x + theta0
```

```python
# Full-Batch Gradient Descent
def full_batch_gradient_descent(x, y, theta0, theta1, learning_rate=0.01, epochs=15):
    loss_history = []
    print("full_batch_gradient_descent")
    for epoch in range(epochs):
        y_pred = model(x, theta0, theta1)
        loss = mse_loss(y_pred, y)
        loss_history.append(loss.item())

        # Backpropagation
        loss.backward()

        # Update parameters
        theta0.data -= learning_rate * theta0.grad
        theta1.data -= learning_rate * theta1.grad

        # Zero gradients
        theta0.grad.zero_()
        theta1.grad.zero_()
        print("epoch", epoch, ": ", loss.item())

    return loss_history
```

```python
# Mini-Batch Gradient Descent -without shuffle
def mini_batch_gradient_descent(x, y, theta0, theta1, batch_size=10, learning_rate=0.01, epochs=15):
    num_samples = x.shape[0]
    loss_history = []
    
    print("mini_batch_gradient_descent")
    
    for epoch in range(epochs):
        # Shuffle the data for each epoch
#         perm = torch.randperm(num_samples)
        for i in range(0, num_samples, batch_size):
#             index = perm[i:i+batch_size]
            x_batch = x[i:i+batch_size]
            y_batch = y[i:i+batch_size]
            
            # Forward pass
            y_pred = model(x_batch, theta0, theta1)
            loss = mse_loss(y_pred, y_batch)
            loss_history.append(loss.item())
            
            # Backward pass
            loss.backward()

            # Update parameters
            theta0.data -= learning_rate * theta0.grad
            theta1.data -= learning_rate * theta1.grad

            # Zero gradients
            theta0.grad.zero_()
            theta1.grad.zero_()
        print("epoch", epoch, ": ", loss.item())
    
    return loss_history
```

```python
# Stochastic Gradient Descent (SGD) - without shuffle
def stochastic_gradient_descent(x, y, theta0, theta1, learning_rate=0.01, epochs=15):
    num_samples = x.shape[0]
    loss_history = []
    print("stochastic_gradient_descent")

    for epoch in range(epochs):
        # Shuffle the data for each epoch
#         perm = torch.randperm(num_samples)
        
        for i in range(num_samples):
#             index = perm[i]
            xi = x[i].view(1, -1)
            yi = y[i].view(1, -1)

            # Forward pass
            y_pred = model(xi, theta0, theta1)
            loss = mse_loss(y_pred, yi)
            loss_history.append(loss.item())

            # Backward pass
            loss.backward()

            # Update parameters
            theta0.data -= learning_rate * theta0.grad
            theta1.data -= learning_rate * theta1.grad

            # Zero gradients
            theta0.grad.zero_()
            theta1.grad.zero_()
        print("epoch", epoch, ": ", loss.item())

    return loss_history


```

### Which optimization process takes a larger number of epochs to converge, and why?

Stochastic Gradient Descent (SGD) updates the parameters more frequently (after each data point) but its updates are noisy. Hence, it takes more iterations to converge, as well as epochs to converge.

Full-Batch Gradient Descent generally takes fewer iterations to converge since it computes the exact gradient using all data points, but it is computationally expensive per iteration.

Mini-Batch Gradient Descent typically offers a good balance between the two, with a smaller batch size providing more frequent updates but less noisy than SGD.

Thus, Stochastic Gradient Descent takes large number of epochs to converge due to: unstable updates, more noises, and more oscillations

### Show the contour plots for different epochs (or show an animation/GIF) for visualisation of optimisation process.

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Set random seed for reproducibility
np.random.seed(45)
num_samples = 40

# Generate data
x1 = np.random.uniform(-1, 1, num_samples)  # Input features
f_x = 3 * x1 + 4                           # True function
eps = np.random.randn(num_samples)          # Noise
y = f_x + eps                               # Noisy target values

# Convert data to PyTorch tensors for automatic differentiation
x1 = torch.tensor(x1, dtype=torch.float32).view(-1, 1)
y = torch.tensor(y, dtype=torch.float32).view(-1, 1)

# Linear regression model
def model(x, theta0, theta1):
    return theta1 * x + theta0

# Mean Squared Error (MSE) loss function
def mse_loss(theta0, theta1, x, y):
    y_pred = model(x, theta0, theta1)
    return torch.mean((y_pred - y) ** 2)

# Initialize parameters for gradient descent
theta0 = torch.tensor([0.0], requires_grad=True)  # Intercept
theta1 = torch.tensor([0.0], requires_grad=True)  # Slope
learning_rate = 0.1
epochs = 100

# Lists to store parameter and loss values
theta0_list = []
theta1_list = []
loss_list = []

# Perform gradient descent
for epoch in range(epochs):
    if theta0.grad is not None:
        theta0.grad.zero_()
    if theta1.grad is not None:
        theta1.grad.zero_()

    # Compute predictions and loss
    loss = mse_loss(theta0, theta1, x1, y)

    # Backpropagate to compute gradients
    loss.backward()

    # Update parameters using gradient descent
    with torch.no_grad():
        theta0 -= learning_rate * theta0.grad
        theta1 -= learning_rate * theta1.grad

    # Store the parameters and loss for visualization
    theta0_list.append(theta0.item())
    theta1_list.append(theta1.item())
    loss_list.append(loss.item())

# 2D Plot: Theta0 vs Theta1, with circles showing updates
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the path of theta updates
ax.plot(theta0_list, theta1_list, label='Theta updates', color='blue', marker='o')

# Labels and title
ax.set_xlabel('Theta 0 (Intercept)')
ax.set_ylabel('Theta 1 (Slope)')
ax.set_title('Updates in Theta 0 and Theta 1 during Gradient Descent')
ax.legend()

# Show the plot
plt.show()

# Create a mesh grid for the loss surface
theta0_vals = np.linspace(-2, 10, 100)
theta1_vals = np.linspace(-2, 10, 100)
theta0_grid, theta1_grid = np.meshgrid(theta0_vals, theta1_vals)

# Compute the loss for each point on the grid
loss_vals = np.zeros_like(theta0_grid)
for i in range(theta0_grid.shape[0]):
    for j in range(theta0_grid.shape[1]):
        theta0_val = torch.tensor(theta0_grid[i, j], dtype=torch.float32)
        theta1_val = torch.tensor(theta1_grid[i, j], dtype=torch.float32)
        loss_vals[i, j] = mse_loss(theta0_val, theta1_val, x1, y).item()

# Set up the figure for loss surface and updates in theta values
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the contour of the loss function
contour = ax.contourf(theta0_grid, theta1_grid, loss_vals, levels=50, cmap='viridis')
plt.colorbar(contour)

# Plot the path of theta updates with circles marking each step
theta_updates = ax.plot([], [], 'ro-', lw=2)[0]  # Path with red dots
current_point = ax.plot([], [], 'bo', color='red', markersize=8)[0]  # Current point (blue circle)

# Set labels and title
ax.set_xlabel('Theta 0 (Intercept)')
ax.set_ylabel('Theta 1 (Slope)')
ax.set_title('Gradient Descent Path on Loss Contour')

# Update function for animation
def update(frame):
    theta_updates.set_data(theta0_list[:frame+1], theta1_list[:frame+1])  # Update path
    current_point.set_data(theta0_list[frame], theta1_list[frame])  # Update current point
    return theta_updates, current_point

# Create animation
ani = FuncAnimation(fig, update, frames=len(theta0_list), interval=200, blit=True)

# Show the animation
plt.show()

# Optionally, save the animation as a GIF
ani.save('theta_updates_animation.gif', writer='imagemagick')
```

```python
import numpy as np
import torch
import matplotlib.pyplot as plt

# Generate some data
np.random.seed(42)
x = np.random.uniform(-1, 1, 40)
y = 3 * x + 4 + np.random.randn(40)

# Convert data to PyTorch tensors
x = torch.tensor(x, dtype=torch.float32).view(-1, 1)
y = torch.tensor(y, dtype=torch.float32).view(-1, 1)

# Linear regression model
def model(x, theta0, theta1):
    return theta1 * x + theta0

# Mean Squared Error loss function
def mse_loss(theta0, theta1, x, y):
    y_pred = model(x, theta0, theta1)
    return torch.mean((y_pred - y) ** 2)

# Initialize parameters
theta0 = torch.tensor([0.0], requires_grad=True)
theta1 = torch.tensor([0.0], requires_grad=True)

# Momentum parameters
alpha = 0.01  # Learning rate
beta = 0.9    # Momentum factor
v_theta0 = torch.tensor([0.0])  # Initial velocity for theta0
v_theta1 = torch.tensor([0.0])  # Initial velocity for theta1

# Number of iterations
iterations = 100

# Lists to store values for plotting
theta0_list = []
theta1_list = []
loss_list = []

# Gradient Descent with Momentum
for i in range(iterations):
    if theta0.grad is not None:
        theta0.grad.zero_()
    if theta1.grad is not None:
        theta1.grad.zero_()

    # Compute the loss
    loss = mse_loss(theta0, theta1, x, y)

    # Perform backpropagation to compute gradients
    loss.backward()

    # Update velocity for theta0 and theta1
    v_theta0 = beta * v_theta0 - alpha * theta0.grad
    v_theta1 = beta * v_theta1 - alpha * theta1.grad

    # Update parameters using momentum
    with torch.no_grad():
        theta0 += v_theta0
        theta1 += v_theta1

    # Store the parameter and loss values
    theta0_list.append(theta0.item())
    theta1_list.append(theta1.item())
    loss_list.append(loss.item())

# Plot the loss over iterations
plt.plot(loss_list)
plt.title("Loss vs Iterations (Gradient Descent with Momentum)")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import torch

# Set random seed for reproducibility
np.random.seed(45)
num_samples = 40

# Generate data for linear regression
x1 = np.random.uniform(-1, 1, num_samples)  # Input features
f_x = 3 * x1 + 4                           # True linear function
eps = np.random.randn(num_samples)          # Noise
y = f_x + eps                               # Target values

# Convert data to PyTorch tensors for automatic differentiation
x1 = torch.tensor(x1, dtype=torch.float32).view(-1, 1)
y = torch.tensor(y, dtype=torch.float32).view(-1, 1)

# Linear regression model
def model(x, theta0, theta1):
    return theta1 * x + theta0

# Mean Squared Error (MSE) cost function
def mse_loss(theta0, theta1, x, y):
    y_pred = model(x, theta0, theta1)
    return torch.mean((y_pred - y) ** 2)

# Initialize parameters for gradient descent
theta0 = torch.tensor([4.1], requires_grad=True)  # Intercept, initial value
theta1 = torch.tensor([4.1], requires_grad=True)  # Slope, initial value
alpha = 0.001  # Learning rate
iterations = 2  # Number of iterations

# Lists to store parameter and loss values
theta0_list = []
theta1_list = []
loss_list = []

# Gradient descent loop with plotting
for i in range(iterations):
    plt.figure(figsize=(8, 6))
    
    # Plot the quadratic cost function for visual comparison
    p_range = np.linspace(0, 8, 1000)
    loss_range = [(mse_loss(torch.tensor(p, dtype=torch.float32), torch.tensor(theta1.item(), dtype=torch.float32), x1, y)).item() for p in p_range]
    
    # Plot the cost function
    plt.plot(p_range, loss_range, label='Cost function: MSE', color='blue')
    
    # Previous values of theta0 and theta1 (before update)
    prev_theta0 = theta0.item()
    prev_theta1 = theta1.item()
    
    # Calculate the loss
    loss = mse_loss(theta0, theta1, x1, y)
    loss_list.append(loss.item())
    
    # Perform backpropagation to compute the gradients
    loss.backward()

    # Update the parameters using gradient descent
    with torch.no_grad():
        theta0 -= alpha * theta0.grad
        theta1 -= alpha * theta1.grad
    
    # Store the updated parameters
    theta0_list.append(theta0.item())
    theta1_list.append(theta1.item())
    
    # Zero the gradients for the next step
    theta0.grad.zero_()
    theta1.grad.zero_()
    
    # Plot the arrow showing the gradient descent step for theta0
    plt.arrow(prev_theta0, mse_loss(torch.tensor(prev_theta0), torch.tensor(prev_theta1), x1, y).item(),
              theta0.item() - prev_theta0, mse_loss(theta0, theta1, x1, y).item() - mse_loss(torch.tensor(prev_theta0), torch.tensor(prev_theta1), x1, y).item(),
              head_width=0.1, head_length=0.5, fc='green', ec='green')

    # Scatter plot to mark the current point (theta0, cost)
    plt.scatter([theta0.item()], [mse_loss(theta0, theta1, x1, y).item()], color='red', s=100, zorder=5)
    
    # Set plot labels and title
    plt.xlabel("Theta0 (Intercept)")
    plt.ylabel("Cost (MSE)")
    plt.title(f"Iteration {i+1} (Learning Rate: {alpha})")

    # Display grid and legend
    plt.grid(True)
    plt.legend()

    # Show the plot
    plt.show()
```
